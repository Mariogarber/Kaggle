```python
import pandas as pd 

data = pd.read_csv('earthquakes_scaled.csv', index_col="index")

data.drop(columns=['Unnamed: 0'], inplace=True)

data
```

```python
import keras as kr

model = kr.models.Sequential()

model.add(kr.layers.Dense(50, input_shape=(20,), activation="relu", kernel_initializer="he_normal"))
model.add(kr.layers.Dense(100, activation="relu", kernel_initializer="he_normal"))
model.add(kr.layers.BatchNormalization())
model.add(kr.layers.Dense(150, activation="relu", kernel_initializer="he_normal"))
model.add(kr.layers.BatchNormalization())
model.add(kr.layers.Dense(150, activation="relu", kernel_initializer="he_normal"))
model.add(kr.layers.Dense(1, activation="linear", kernel_initializer="he_normal"))

model.compile(kr.optimizers.Adam(), loss=kr.losses.MeanAbsoluteError(), metrics=[kr.metrics.MeanAbsoluteError(), kr.metrics.MeanSquaredError()])
```

```python
#I want to split the data into training and testing data

from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.10)

train_data.isna().sum()
```

```python
X_train = train_data.drop(columns=['magnitude'])
y_train = train_data['magnitude']

model.fit(X_train, y_train, batch_size=48, epochs=1000)

X_test = test_data.drop(columns=['magnitude'])
y_test = test_data['magnitude']

model.evaluate(X_test, y_test)
```
