```python
import numpy as np
import pandas as pd

df = pd.read_csv('earthquakes.csv', index_col="id", parse_dates=True)

df
```

```python
df["magType"].value_counts()
```

```python
data = df.drop(["title", "url", "detailUrl", "what3words", "postcode", "locationDetails", "type", "alert", "status", "place", "placeOnly", "net", "ids", "geometryType", "types"], axis=1)

print(data)
```

```python
list_objects = []

for column in data.columns:
    if data[column].dtype == object:
        print(column, data[column].dtype)
        list_objects.append(column)
```

```python
data[list_objects]
```

```python
data = data.drop(["code", "sources", "locality", "location", "date", "timezone"], axis=1)
list_objects = ["magType", "continent", "country", "subnational", "city"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in list_objects:
    data[column] = le.fit_transform(data[column])
```

```python
data
```

```python
#I want to apply pca to my dataset and show the explained variance ratio of each component

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

pca = PCA()
pca.fit(data_scaled)

explained_variance = pca.explained_variance_ratio_

explained_variance
```

```python
variance_sum = 0
variance_list = []
for i in range(len(explained_variance)):
    variance_sum += explained_variance[i]
    variance_list.append(variance_sum)
    print(f"Explained variance of component {i}: {explained_variance[i]}")
    print(f"Total explained variance until component {i}: {variance_sum}")
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, len(variance_list) + 1), variance_list, marker="o")
plt.plot(range(1, len(variance_list) + 1), list(map(lambda x: x*0+0.9, variance_list)))
plt.plot(range(1, len(variance_list) + 1), list(map(lambda x: x*0+0.95, variance_list)))
plt.xlabel("Number of Components")
plt.ylabel("Explained Variance")
```

```python
#I want to study atipics points in my dataset

from sklearn.ensemble import IsolationForest
'''
NO TENEMOS INFORMACIÓN DEL PORCENTAJE DE OUTLIERS EN LOS DATOS, CONTAMINATION = AUTO
ISOLATION FOREST CREA UN MODELO DE BOSQUE ALEATORIO Y DETECTA OUTLIERS EN FUNCION
DE LO FÁCIL QUE ES SEPARARLOS DE LOS DEMÁS DATOS

ELIMINA MÁS DE UN 10% DE LOS DATOS, POR LO QUE NO ES UNA BUENA IDEA UTILIZAR EL AUTO
'''


clf = IsolationForest(n_estimators= 300, contamination=0.01)
clf.fit(data)

outliers = clf.predict(data)

outliers

data["outliers"] = outliers

data[data["outliers"] == -1]
```

```python
df_outliers = pd.DataFrame(columns=["mean_outliers", "mean_no_outliers"])



df_outliers["mean_outliers"] = data[(data["outliers"] == 1)].mean() 
df_outliers["mean_no_outliers"] = data[(data["outliers"] == -1)].mean()

df_outliers["difference"] = df_outliers.apply(lambda x: abs(x["mean_outliers"] - x["mean_no_outliers"]), axis=1)

df_outliers
```

```python
from scipy import stats

stat_list = []
pvalue_list = []

for column in data.columns:
    stat, pvalue = stats.kstest(data[column], "norm")
    stat_list.append(stat)
    pvalue_list.append(pvalue)
    print(f"Column: {column} - Statistic: {stat} - P-value: {pvalue}")

df_ks = pd.DataFrame(columns=["statistic", "pvalue"], index=data.columns)

df_ks[df_ks["pvalue"] > 0.05]
```

```python
from scipy import stats

stat_list = []
pvalue_list = []

for column in data.columns:
    stat, pvalue = stats.shapiro(data[column])
    stat_list.append(stat)
    pvalue_list.append(pvalue)
    print(f"Column: {column} - Statistic: {stat} - P-value: {pvalue}")

df_shap = pd.DataFrame(columns=["statistic", "pvalue"], index=data.columns)

df_shap[df_shap["pvalue"] > 0.05]
```
